```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['sale_date'] = pd.to_datetime(df['sale_date'])
```

```python
df['estimated_profit'] = df['revenue_usd'] * 0.30
```

```python
df['month'] = df['sale_date'].dt.month_name()
```

```python
sns.histplot(df['revenue_usd'])
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
df.groupby('category')['revenue_usd'].sum().sort_values().plot(kind='barh')

plt.show()
```

```python
monthly_sales = df.groupby('month')['revenue_usd'].sum()

monthly_sales.plot(kind='line')

plt.show()
```

```python
sns.scatterplot(
    x='customer_rating',
    y='revenue_usd',
    data=df
)

plt.show()
```

```python
sns.countplot(x='return_status', data=df)

plt.show()
```
